```python
import pandas as pd
import numpy as np
from rapidfuzz import fuzz
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
import openpyxl
from openpyxl.drawing.image import Image

# ============================================================
# 1. ЗАГРУЗКА ДАННЫХ
# ============================================================

file_path = "dataset_stepstone-scraper-fast-reliable-4-1k_2025-12-15_19-18-23-513.xlsx"
df = pd.read_excel(file_path)

# ============================================================
# 2. СПИСОК ПОИСКОВЫХ ТИТЛОВ
# ============================================================

search_titles = [
    "Business Transformation Analyst",
    "Digital Process Analyst",
    "AI Governance Analyst",
    "AI Automation Specialist",
    "Prompt Engineer",
    "Junior Automation Specialist",
    "AI Project Manager",
    "AI Product Manager",
    "Prozessmanager/in",
    "KI-Manager/in",
    "Prozessmanager/in – RPA",
    "KI-Prompter",
    "Prozessmanager/in – RPA (Junior)"
]

# ============================================================
# 3. СИНОНИМЫ
# ============================================================

synonyms = {
    "AI Product Manager": [
        "product owner ai", "ai produktmanager", "ml product manager",
        "product manager ai", "ai po", "ai product owner",
        "product manager digital", "product manager data", "product manager tech"
    ],
    "Prompt Engineer": [
        "llm engineer", "nlp engineer", "ai content engineer",
        "generative ai engineer", "prompt specialist",
        "prompting", "llm", "genai", "chatgpt", "ai text"
    ],
    "Prozessmanager/in": [
        "process manager", "business process manager", "process owner",
        "prozess spezialist", "prozessmanager", "process analyst",
        "business analyst process", "process consultant",
        "portfolio process analyst", "process governance analyst"
    ],
    "AI Automation Specialist": [
        "rpa developer", "automation engineer", "intelligent automation",
        "rpa specialist", "automation specialist",
        "low-code automation", "n8n", "make.com", "workflow automation",
        "facility automation", "building automation"
    ],
    "Junior Automation Specialist": [
        "junior rpa", "junior automation", "junior process automation",
        "junior workflow automation"
    ],
    "KI-Manager/in": [
        "ai manager", "manager ki", "leiter ki", "head of ai",
        "ai lead", "ai team lead"
    ],
    "AI Project Manager": [
        "project manager ai", "ai projektmanager", "ml project manager",
        "project manager digital", "project manager data"
    ],
    "Digital Process Analyst": [
        "digital analyst", "web analyst", "data analyst",
        "digital data analyst", "digital forensic analyst",
        "strategic analyst digital", "junior analyst digital",
        "digital retail analyst", "digital feature analyst"
    ],
    "Business Transformation Analyst": [
        "business analyst", "business analyst digital",
        "business analyst insurance", "business analyst finance",
        "business analyst crm", "business analyst growth",
        "business analyst camunda", "business analyst supply chain",
        "business analyst servicecontrolling"
    ],
    "AI Governance Analyst": [
        "ai compliance", "ai governance", "responsible ai",
        "digital compliance", "regulatory compliance",
        "governance analyst", "process governance"
    ],
    "KI-Prompter": [
        "prompting specialist", "ki prompt", "ai prompt",
        "prompt writer", "prompt designer"
    ],
    "Prozessmanager/in – RPA": [
        "rpa process manager", "rpa prozessmanager",
        "process automation rpa", "rpa analyst"
    ],
    "Prozessmanager/in – RPA (Junior)": [
        "junior rpa process manager", "junior rpa prozessmanager",
        "junior rpa analyst"
    ]
}

# ============================================================
# 4. КЛЮЧЕВЫЕ СЛОВА
# ============================================================

keywords = {
    "AI Product Manager": ["product", "manager", "ai", "ml", "digital product", "product owner"],
    "Prompt Engineer": ["prompt", "llm", "nlp", "genai", "chatgpt", "ai text", "prompting"],
    "Prozessmanager/in": ["process", "prozess", "workflow", "governance", "operations",
                          "process optimization", "process improvement"],
    "AI Automation Specialist": ["automation", "rpa", "workflow", "low-code", "n8n", "make.com",
                                 "intelligent automation", "robotic"],
    "Junior Automation Specialist": ["junior", "automation", "rpa", "workflow"],
    "KI-Manager/in": ["ai", "ki", "manager", "lead", "head"],
    "AI Project Manager": ["project", "manager", "ai", "digital project", "it project"],
    "Digital Process Analyst": ["digital", "analyst", "web", "data", "forensic",
                                "dashboard", "tracking", "kpi", "reporting"],
    "Business Transformation Analyst": ["business", "analyst", "consultant", "requirements",
                                        "stakeholder", "process", "insurance", "finance"],
    "AI Governance Analyst": ["governance", "compliance", "regulatory", "responsible ai",
                              "audit", "policy"],
    "KI-Prompter": ["prompt", "ki", "ai", "genai"],
    "Prozessmanager/in – RPA": ["rpa", "automation", "prozess", "robotic"],
    "Prozessmanager/in – RPA (Junior)": ["junior", "rpa", "prozess", "automation"]
}

# ============================================================
# 5. ФУНКЦИЯ МЯГКОЙ КЛАССИФИКАЦИИ
# ============================================================

def classify_soft(row):
    """
    Выполняет мягкую классификацию строки вакансии:
    fuzzy + синонимы + ключевые слова.
    Возвращает серию с метриками.
    """
    text = f"{row.get('title', '')} {row.get('description', '')}".lower()

    scores = []
    fuzzy_scores = []
    synonym_scores = []
    keyword_scores = []
    matched_synonyms = []

    for st in search_titles:

        # fuzzy score
        fuzzy_score = max(
            fuzz.partial_ratio(text, st.lower()),
            fuzz.token_set_ratio(text, st.lower()),
            fuzz.WRatio(text, st.lower())
        ) / 100

        # synonym score
        syn_hits = [s for s in synonyms.get(st, []) if s in text]
        syn_score = len(syn_hits)

        # keyword score
        kw_score = sum(1 for kw in keywords.get(st, []) if kw in text)

        # weighted final score
        final_score = 0.5 * fuzzy_score + 0.3 * syn_score + 0.2 * kw_score

        scores.append(final_score)
        fuzzy_scores.append(fuzzy_score)
        synonym_scores.append(syn_score)
        keyword_scores.append(kw_score)
        matched_synonyms.append(syn_hits)

    best_idx = int(np.argmax(scores))

    return pd.Series({
        "search_title": search_titles[best_idx],
        "final_score": scores[best_idx],
        "fuzzy_score": fuzzy_scores[best_idx],
        "synonym_score": synonym_scores[best_idx],
        "keyword_score": keyword_scores[best_idx],
        "matched_synonyms": matched_synonyms[best_idx]
    })

# ============================================================
# 6. ПРИМЕНЕНИЕ КЛАССИФИКАЦИИ
# ============================================================

tqdm.pandas(desc="Классификация строк")
df_results = df.progress_apply(classify_soft, axis=1)

# удаляем старые служебные колонки
cols_to_remove = [
    "search_title", "final_score", "fuzzy_score",
    "synonym_score", "keyword_score", "matched_synonyms"
]
df = df.drop(columns=[c for c in cols_to_remove if c in df.columns], errors="ignore")

# добавляем новые результаты
df = pd.concat([df, df_results], axis=1)

# ============================================================
# 7. РАСПРЕДЕЛЕНИЕ
# ============================================================

dist = df["search_title"].value_counts().rename_axis("search_title").reset_index(name="count")
dist["percent"] = (dist["count"] / len(df) * 100).round(2)

# ============================================================
# 8. ТОП СИНОНИМОВ
# ============================================================

synonym_exploded = df.explode("matched_synonyms")
synonym_counts = synonym_exploded["matched_synonyms"].value_counts()

# ============================================================
# 9. НИЗКАЯ УВЕРЕННОСТЬ
# ============================================================

low_conf = df.sort_values("final_score").head(50)

# ============================================================
# 10. ВЫВОД ТАБЛИЦ НА ЭКРАН
# ============================================================

print("\n=== DISTRIBUTION BY search_title ===")
display(dist)

print("\n=== TOP SYNONYM MATCHES (top 20) ===")
display(synonym_counts.head(20))

print("\n=== LOW CONFIDENCE ROWS (top ) ===")
display(low_conf[["title", "search_title", "final_score", "fuzzy_score", "synonym_score", "keyword_score"]])

# ============================================================
# 11. ВИЗУАЛИЗАЦИИ SEABORN
# ============================================================

# ---------- 11.1 Barplot распределения search_title ----------
plt.figure(figsize=(14, 6))

# seaborn ≥0.14 требует hue= для palette
sns.barplot(
    data=dist,
    x="search_title",
    y="count",
    hue="search_title",
    palette="Blues_d",
    legend=False
)

# подписи значений
for i, v in enumerate(dist["count"]):
    plt.text(i, v + 1, str(v), ha="center", va="bottom", fontsize=9)

plt.xticks(rotation=75, ha="right")
plt.title("Distribution of search_title")
plt.tight_layout()
plt.savefig("plot_distribution.png")
plt.show()
plt.close()

# ---------- 11.2 Barplot топ-синонимов ----------
top_syn = synonym_counts.head(20)

plt.figure(figsize=(10, 8))

sns.barplot(
    x=top_syn.values,
    y=top_syn.index.astype(str),
    hue=top_syn.index.astype(str),
    palette="Greens_d",
    legend=False
)

# подписи значений
for i, v in enumerate(top_syn.values):
    plt.text(v + 0.2, i, str(v), va="center", fontsize=9)

plt.title("Top Synonym Matches")
plt.xlabel("Count")
plt.ylabel("Synonym")
plt.tight_layout()
plt.savefig("plot_synonyms.png")
plt.show()
plt.close()

# ---------- 11.3 Scatterplot уверенности ----------
plt.figure(figsize=(12, 5))
sns.scatterplot(x=df.index, y=df["final_score"], alpha=0.6)
plt.title("Final Score Confidence Scatter")
plt.xlabel("Row index")
plt.ylabel("Final score")
plt.grid(True, linestyle="--", alpha=0.4)
plt.tight_layout()
plt.savefig("plot_confidence.png")
plt.show()
plt.close()

# ---------- 11.4 Heatmap корреляций ----------
corr_df = df[["final_score", "fuzzy_score", "synonym_score", "keyword_score"]]

plt.figure(figsize=(8, 6))
sns.heatmap(corr_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.savefig("plot_heatmap.png")
plt.show()
plt.close()

# ============================================================
# 12. СОХРАНЕНИЕ В EXCEL С ВКЛАДКАМИ
# ============================================================

with pd.ExcelWriter("Fuzzy_classified_output.xlsx", engine="openpyxl") as writer:
    df.to_excel(writer, sheet_name="Full", index=False)
    dist.to_excel(writer, sheet_name="search_title distribution", index=False)
    synonym_counts.to_excel(writer, sheet_name="SYNONYM", index=True)
    low_conf.to_excel(writer, sheet_name="LOW CONFIDENCE ROWS", index=False)

    workbook = writer.book

    # вкладка: Distribution plot
    ws1 = workbook.create_sheet("Plot Distribution")
    ws1.add_image(Image("plot_distribution.png"), "A1")

    # вкладка: Synonym plot
    ws2 = workbook.create_sheet("Plot Synonyms")
    ws2.add_image(Image("plot_synonyms.png"), "A1")

    # вкладка: Confidence plot
    ws3 = workbook.create_sheet("Plot Confidence")
    ws3.add_image(Image("plot_confidence.png"), "A1")

    # вкладка: Heatmap
    ws4 = workbook.create_sheet("Plot Heatmap")
    ws4.add_image(Image("plot_heatmap.png"), "A1")

print("\nГотово. Файл сохранён: Fuzzy_classified_output.xlsx")
```

```python
# ============================================================
# IMPORTS
# ============================================================

import pandas as pd
import numpy as np
from rapidfuzz import fuzz
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
import openpyxl
from openpyxl.drawing.image import Image


# ============================================================
# 1. LOAD DATA
# ============================================================

file_path = "dataset_stepstone-scraper-fast-reliable-4-1k_2025-12-15_19-18-23-513.xlsx"
df = pd.read_excel(file_path)


# ============================================================
# 2. SEARCH TITLES, SYNONYMS, KEYWORDS
# ============================================================

search_titles = [
    "Business Transformation Analyst",
    "Digital Process Analyst",
    "AI Governance Analyst",
    "AI Automation Specialist",
    "Prompt Engineer",
    "Junior Automation Specialist",
    "AI Project Manager",
    "AI Product Manager",
    "Prozessmanager/in",
    "KI-Manager/in",
    "Prozessmanager/in – RPA",
    "KI-Prompter",
    "Prozessmanager/in – RPA (Junior)"
]

synonyms = {
    "AI Product Manager": [
        "product owner ai", "ai produktmanager", "ml product manager",
        "product manager ai", "ai po", "ai product owner",
        "product manager digital", "product manager data", "product manager tech"
    ],
    "Prompt Engineer": [
        "llm engineer", "nlp engineer", "ai content engineer",
        "generative ai engineer", "prompt specialist",
        "prompting", "llm", "genai", "chatgpt", "ai text"
    ],
    "Prozessmanager/in": [
        "process manager", "business process manager", "process owner",
        "prozess spezialist", "prozessmanager", "process analyst",
        "business analyst process", "process consultant",
        "portfolio process analyst", "process governance analyst"
    ],
    "AI Automation Specialist": [
        "rpa developer", "automation engineer", "intelligent automation",
        "rpa specialist", "automation specialist",
        "low-code automation", "n8n", "make.com", "workflow automation",
        "facility automation", "building automation"
    ],
    "Junior Automation Specialist": [
        "junior rpa", "junior automation", "junior process automation",
        "junior workflow automation"
    ],
    "KI-Manager/in": [
        "ai manager", "manager ki", "leiter ki", "head of ai",
        "ai lead", "ai team lead"
    ],
    "AI Project Manager": [
        "project manager ai", "ai projektmanager", "ml project manager",
        "project manager digital", "project manager data"
    ],
    "Digital Process Analyst": [
        "digital analyst", "web analyst", "data analyst",
        "digital data analyst", "digital forensic analyst",
        "strategic analyst digital", "junior analyst digital",
        "digital retail analyst", "digital feature analyst"
    ],
    "Business Transformation Analyst": [
        "business analyst", "business analyst digital",
        "business analyst insurance", "business analyst finance",
        "business analyst crm", "business analyst growth",
        "business analyst camunda", "business analyst supply chain",
        "business analyst servicecontrolling"
    ],
    "AI Governance Analyst": [
        "ai compliance", "ai governance", "responsible ai",
        "digital compliance", "regulatory compliance",
        "governance analyst", "process governance"
    ],
    "KI-Prompter": [
        "prompting specialist", "ki prompt", "ai prompt",
        "prompt writer", "prompt designer"
    ],
    "Prozessmanager/in – RPA": [
        "rpa process manager", "rpa prozessmanager",
        "process automation rpa", "rpa analyst"
    ],
    "Prozessmanager/in – RPA (Junior)": [
        "junior rpa process manager", "junior rpa prozessmanager",
        "junior rpa analyst"
    ]
}

keywords = {
    "AI Product Manager": ["product", "manager", "ai", "ml", "digital product", "product owner"],
    "Prompt Engineer": ["prompt", "llm", "nlp", "genai", "chatgpt", "ai text", "prompting"],
    "Prozessmanager/in": ["process", "prozess", "workflow", "governance", "operations",
                          "process optimization", "process improvement"],
    "AI Automation Specialist": ["automation", "rpa", "workflow", "low-code", "n8n", "make.com",
                                 "intelligent automation", "robotic"],
    "Junior Automation Specialist": ["junior", "automation", "rpa", "workflow"],
    "KI-Manager/in": ["ai", "ki", "manager", "lead", "head"],
    "AI Project Manager": ["project", "manager", "ai", "digital project", "it project"],
    "Digital Process Analyst": ["digital", "analyst", "web", "data", "forensic",
                                "dashboard", "tracking", "kpi", "reporting"],
    "Business Transformation Analyst": ["business", "analyst", "consultant", "requirements",
                                        "stakeholder", "process", "insurance", "finance"],
    "AI Governance Analyst": ["governance", "compliance", "regulatory", "responsible ai",
                              "audit", "policy"],
    "KI-Prompter": ["prompt", "ki", "ai", "genai"],
    "Prozessmanager/in – RPA": ["rpa", "automation", "prozess", "robotic"],
    "Prozessmanager/in – RPA (Junior)": ["junior", "rpa", "prozess", "automation"]
}


# ============================================================
# 3. EXPERIENCE CLASSIFIER
# ============================================================

def extract_experience(text, employment_type=None):
   def extract_experience(text, employment_type=None):
    """
    Определяет требуемый уровень опыта для вакансии.

    Логика работы:
    1. Анализирует текст вакансии (title + description):
       - Если встречаются признаки студенческих/учебных позиций 
         (Werkstudent, Praktikum, Ausbildung, Studium, Trainee и т.п.) → "no_experience".
       - Если встречаются фразы про первые профессиональные навыки 
         ("erste Erfahrungen", "erste Berufserfahrung", "mindestens 1 Jahr") → "1-2_years".
       - Если встречаются фразы про многолетний опыт 
         ("mehrjährige Berufserfahrung", "mindestens 3 Jahre", "5 Jahre") → "3+_years".

    2. Если в тексте нет признаков опыта, анализирует поле employmentType:
       - Если оно относится к учебным/студенческим категориям 
         (Ausbildung, Studium, Praktikum, Werkstudent, Trainee) → "no_experience".
       - "Feste Anstellung" НЕ считается признаком опыта → возвращает "unknown".

    3. Если ни текст, ни employmentType не дают информации → "unknown".

    Параметры:
        text (str): текст вакансии (title + description).
        employment_type (str): значение поля employmentType.

    Возвращает:
        str: одна из категорий:
             "no_experience", "1-2_years", "3+_years", "unknown".
    """

    
    # 1. Анализ текста вакансии
    if not isinstance(text, str):
        text = ""
    t = text.lower()

    # 0 — опыт не требуется
    no_exp_patterns = [
        "werkstudent", "praktikum", "ausbildung", "studium",
        "duales studium", "trainee", "berufseinstieg",
        "studentenjob", "studentenjobs", "werkstudenten",
        "bachelor-/master-/diplom-arbeiten"
    ]
    if any(p in t for p in no_exp_patterns):
        return "no_experience"

    # 1–2 года опыта
    mid_exp_patterns = [
        "erste erfahrungen", "erste berufserfahrung",
        "erste praktische erfahrungen",
        "mindestens einjährige", "mindestens 1 jahr",
        "1 jahr", "1-2 jahre", "1 bis 2 jahre"
    ]
    if any(p in t for p in mid_exp_patterns):
        return "1-2_years"

    # 3+ лет опыта
    high_exp_patterns = [
        "mehrjährige", "mehrere jahre", "mindestens 3 jahre",
        "mindestens drei jahre", "mindestens 5 jahre",
        "5 jahre", "langjährige"
    ]
    if any(p in t for p in high_exp_patterns):
        return "3+_years"


    # 2. Анализ employmentType
    if isinstance(employment_type, str):
        et = employment_type.lower()

        # категории без опыта
        if any(p in et for p in no_exp_patterns):
            return "no_experience"

        # trainee → no experience
        if "trainee" in et:
            return "no_experience"

        # Ausbildung, Studium → no experience
        if "ausbildung" in et or "studium" in et:
            return "no_experience"

        # Werkstudent → no experience
        if "werkstudent" in et:
            return "no_experience"

        # Praktikum → no experience
        if "praktikum" in et:
            return "no_experience"

        # Feste Anstellung → НЕ опыт → вернуть unknown
        if "feste anstellung" in et:
            return "unknown"

    
    # 3. Если ничего не найдено    
    return "unknown"




# ============================================================
# 4. SOFT CLASSIFIER FOR SEARCH TITLES
# ============================================================

def classify_soft(row):
    def classify_soft(row):
    """
    Выполняет мягкую классификацию вакансии по одному из заранее заданных search_title.

    Логика работы:
    1. Объединяет title и description в единый текст и приводит его к нижнему регистру.
    2. Для каждого search_title вычисляет три независимых сигнала:
       
       • fuzzy_score — степень текстового сходства между вакансией и названием search_title
         (используются partial_ratio, token_set_ratio и WRatio из RapidFuzz).

       • synonym_score — количество совпавших синонимов, специфичных для данного search_title.

       • keyword_score — количество совпавших ключевых слов, характерных для search_title.

    3. Комбинирует сигналы в итоговый final_score по формуле:
           final_score = 0.5*fuzzy + 0.3*synonyms + 0.2*keywords

       Такой подход делает классификацию мягкой и устойчивой:
       fuzzy даёт общий смысл, синонимы усиливают точность, ключевые слова добавляют контекст.

    4. Выбирает search_title с максимальным final_score.

    Возвращает:
        pandas.Series с полями:
            - search_title: выбранная категория
            - final_score: итоговый взвешенный скор
            - fuzzy_score: степень текстового сходства
            - synonym_score: количество совпавших синонимов
            - keyword_score: количество совпавших ключевых слов
            - matched_synonyms: список найденных синонимов

    Функция обеспечивает прозрачную и объяснимую классификацию,
    позволяя анализировать вклад каждого сигнала в итоговое решение.
    """
    
    text = f"{row.get('title', '')} {row.get('description', '')}".lower()

    scores = []
    fuzzy_scores = []
    synonym_scores = []
    keyword_scores = []
    matched_synonyms = []

    for st in search_titles:

        fuzzy_score = max(
            fuzz.partial_ratio(text, st.lower()),
            fuzz.token_set_ratio(text, st.lower()),
            fuzz.WRatio(text, st.lower())
        ) / 100

        syn_hits = [s for s in synonyms.get(st, []) if s in text]
        syn_score = len(syn_hits)

        kw_score = sum(1 for kw in keywords.get(st, []) if kw in text)

        final_score = 0.5 * fuzzy_score + 0.3 * syn_score + 0.2 * kw_score

        scores.append(final_score)
        fuzzy_scores.append(fuzzy_score)
        synonym_scores.append(syn_score)
        keyword_scores.append(kw_score)
        matched_synonyms.append(syn_hits)

    best_idx = int(np.argmax(scores))

    return pd.Series({
        "search_title": search_titles[best_idx],
        "final_score": scores[best_idx],
        "fuzzy_score": fuzzy_scores[best_idx],
        "synonym_score": synonym_scores[best_idx],
        "keyword_score": keyword_scores[best_idx],
        "matched_synonyms": matched_synonyms[best_idx]
    })


# ============================================================
# 5. APPLY CLASSIFIERS
# ============================================================

tqdm.pandas(desc="Classifying search titles")
df_results = df.progress_apply(classify_soft, axis=1)

df = pd.concat([df, df_results], axis=1)

#df["necessary_experience"] = df["employmentType"].apply(extract_experience)
#df["necessary_experience"] = (df["title"].astype(str) + " " + df["description"].astype(str)).apply(extract_experience)
df["necessary_experience"] = df.apply(
    lambda row: extract_experience(
        text=f"{row.get('title', '')} {row.get('description', '')}",
        employment_type=row.get("employmentType", "")
    ),
    axis=1
)


# ============================================================
# 6. DISTRIBUTIONS
# ============================================================

dist_titles = df["search_title"].value_counts().rename_axis("search_title").reset_index(name="count")
dist_titles["percent"] = (dist_titles["count"] / len(df) * 100).round(2)

dist_exp = df["necessary_experience"].value_counts().rename_axis("experience").reset_index(name="count")
dist_exp["percent"] = (dist_exp["count"] / len(df) * 100).round(2)


# ============================================================
# 7. VISUALIZATIONS
# ============================================================

def plot_bar(df_plot, x, y, title, filename, palette):
    plt.figure(figsize=(12, 6))
    sns.barplot(data=df_plot, x=x, y=y, hue=x, palette=palette, legend=False)

    for i, v in enumerate(df_plot[y]):
        plt.text(i, v + 1, str(v), ha="center", fontsize=9)

    plt.title(title)
    plt.xticks(rotation=75, ha="right")
    plt.tight_layout()
    plt.savefig(filename)
    plt.show()
    plt.close()


plot_bar(dist_titles, "search_title", "count", "Distribution of Search Titles", "plot_titles.png", "Blues_d")
plot_bar(dist_exp, "experience", "count", "Distribution of Experience", "plot_experience.png", "Purples_d")


# ============================================================
# 8. EXPORT TO EXCEL
# ============================================================

with pd.ExcelWriter("Fuzzy_classified_output.xlsx", engine="openpyxl") as writer:
    df.to_excel(writer, sheet_name="Full", index=False)
    dist_titles.to_excel(writer, sheet_name="search_title distribution", index=False)
    dist_exp.to_excel(writer, sheet_name="experience distribution", index=False)

    workbook = writer.book

    ws1 = workbook.create_sheet("Plot Titles")
    ws1.add_image(Image("plot_titles.png"), "A1")

    ws2 = workbook.create_sheet("Plot Experience")
    ws2.add_image(Image("plot_experience.png"), "A1")

print("\nГотово. Файл сохранён: Fuzzy_classified_output.xlsx")
```
