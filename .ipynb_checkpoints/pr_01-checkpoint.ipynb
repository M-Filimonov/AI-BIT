```python
import pandas as pd
import os

# –ü—É—Ç—å –∫ –ø–∞–ø–∫–µ —Å –¥–∞–Ω–Ω—ã–º–∏
data_folder = "data"

# –ù–∞–∑–≤–∞–Ω–∏—è —Ñ–∞–π–ª–æ–≤
file1 = "01_E-Commerce_dataset_linkedin-jobs-scraper-api_2025-10-27_16-57-38-442.xlsx"
file2 = "01_Marketplace Specialist_dataset_linkedin-jobs-scraper_2025-10-19_15-38-48-104.xlsx"

# –ü–æ–ª–Ω—ã–µ –ø—É—Ç–∏
path1 = os.path.join(data_folder, file1)
path2 = os.path.join(data_folder, file2)

# –ó–∞–≥—Ä—É–∑–∫–∞ —Ñ–∞–π–ª–æ–≤
df1 = pd.read_excel(path1)
df2 = pd.read_excel(path2)

# –ü—Ä–∏–≤–µ–¥–µ–Ω–∏–µ –Ω–∞–∑–≤–∞–Ω–∏–π —Å—Ç–æ–ª–±—Ü–æ–≤ –∫ –Ω–∏–∂–Ω–µ–º—É —Ä–µ–≥–∏—Å—Ç—Ä—É
df1.columns = df1.columns.str.lower()
df2.columns = df2.columns.str.lower()

# –°—Ä–∞–≤–Ω–µ–Ω–∏–µ —Å—Ç—Ä—É–∫—Ç—É—Ä
common_cols = sorted(set(df1.columns).intersection(set(df2.columns)))
unique_df1 = sorted(set(df1.columns) - set(df2.columns))
unique_df2 = sorted(set(df2.columns) - set(df1.columns))

print("‚úÖ –û–±—â–∏–µ –ø–æ–ª—è:", common_cols)
print("üìÅ –£–Ω–∏–∫–∞–ª—å–Ω—ã–µ –ø–æ–ª—è –≤ –ø–µ—Ä–≤–æ–º —Ñ–∞–π–ª–µ:", unique_df1)
print("üìÅ –£–Ω–∏–∫–∞–ª—å–Ω—ã–µ –ø–æ–ª—è –≤–æ –≤—Ç–æ—Ä–æ–º —Ñ–∞–π–ª–µ:", unique_df2)

# –ü—Ä–∏–≤–µ–¥–µ–Ω–∏–µ –∫ –æ–±—â–µ–º—É —Ñ–æ—Ä–º–∞—Ç—É
df1_common = df1[common_cols]
df2_common = df2[common_cols]

# –û–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ
combined_df = pd.concat([df1_common, df2_common], ignore_index=True)

# –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –æ–±—ä–µ–¥–∏–Ω—ë–Ω–Ω–æ–≥–æ —Ñ–∞–π–ª–∞
output_path = os.path.join(data_folder, "combined_ecommerce_jobs.xlsx")
combined_df.to_excel(output_path, index=False)

print(f"‚úÖ –û–±—ä–µ–¥–∏–Ω—ë–Ω–Ω—ã–π —Ñ–∞–π–ª —Å–æ—Ö—Ä–∞–Ω—ë–Ω: {output_path}")
```

```python
import pandas as pd

# –ó–∞–≥—Ä—É–∑–∫–∞ –æ–±—ä–µ–¥–∏–Ω—ë–Ω–Ω–æ–≥–æ —Ñ–∞–π–ª–∞
df = pd.read_excel("data/combined_ecommerce_jobs.xlsx")

# –ü—Ä–∏–≤–µ–¥–µ–Ω–∏–µ –Ω–∞–∑–≤–∞–Ω–∏–π —Å—Ç–æ–ª–±—Ü–æ–≤ –∫ –Ω–∏–∂–Ω–µ–º—É —Ä–µ–≥–∏—Å—Ç—Ä—É
df.columns = df.columns.str.lower()

# –û–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ –Ω–∞–∑–≤–∞–Ω–∏—è –∏ –æ–ø–∏—Å–∞–Ω–∏—è –≤–∞–∫–∞–Ω—Å–∏–∏ –≤ –æ–¥–∏–Ω —Ç–µ–∫—Å—Ç
df["full_text"] = df["job_title"].fillna("") + " " + df["description"].fillna("")
df["full_text"] = df["full_text"].str.lower()

# –ö–ª—é—á–µ–≤—ã–µ —Å–ª–æ–≤–∞ –ø–æ –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏—è–º
keywords_marketplace = ["amazon", "ebay", "otto", "zalando", "marketplace", "marktplatz"]
keywords_sales = ["e-commerce", "webshop", "verkauf", "sales", "umsatz", "direct-to-consumer", "d2c"]
keywords_marketing = ["seo", "sea", "google ads", "meta ads", "performance marketing", "social media", "kampagne", "content", "newsletter"]

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è —Ñ–∏–ª—å—Ç—Ä–∞—Ü–∏–∏ –ø–æ –∫–ª—é—á–µ–≤—ã–º —Å–ª–æ–≤–∞–º
def contains_keywords(text, keywords):
    return any(keyword in text for keyword in keywords)

# –ü—Ä–∏–º–µ–Ω–µ–Ω–∏–µ —Ñ–∏–ª—å—Ç—Ä–æ–≤
df["is_marketplace"] = df["full_text"].apply(lambda x: contains_keywords(x, keywords_marketplace))
df["is_sales"] = df["full_text"].apply(lambda x: contains_keywords(x, keywords_sales))
df["is_marketing"] = df["full_text"].apply(lambda x: contains_keywords(x, keywords_marketing))

# –°–æ–∑–¥–∞–Ω–∏–µ –∫–æ–ª–æ–Ω–∫–∏ —Å –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–µ–º
def assign_direction(row):
    directions = []
    if row["is_marketplace"]:
        directions.append("Marketplace")
    if row["is_sales"]:
        directions.append("Sales")
    if row["is_marketing"]:
        directions.append("Marketing")
    return ", ".join(directions) if directions else "Uncategorized"

df["direction"] = df.apply(assign_direction, axis=1)

# –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –æ—Ç—Ñ–∏–ª—å—Ç—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ —Ñ–∞–π–ª–∞
df.to_excel("data/filtered_ecommerce_jobs.xlsx", index=False)

print("‚úÖ –§–∏–ª—å—Ç—Ä–∞—Ü–∏—è –∑–∞–≤–µ—Ä—à–µ–Ω–∞. –î–æ–±–∞–≤–ª–µ–Ω–∞ –∫–æ–ª–æ–Ω–∫–∞ 'direction'.")
```

# Python-—Å–∫—Ä–∏–ø—Ç, –æ–±—ä–µ–¥–∏–Ω—è—é—â–∏–π –≤—Å–µ –∫–ª—é—á–µ–≤—ã–µ —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞ –≤–∞–∫–∞–Ω—Å–∏–π:
- –ö–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏—è –ø–æ –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏—é: Marketplace, Marketing, Sales
- –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ: eCommerce –∏–ª–∏ offline
- –ò–∑–≤–ª–µ—á–µ–Ω–∏–µ: —Ö–∞—Ä–¥ —Å–∫–∏–ª–ª–æ–≤, —Å–æ—Ñ—Ç —Å–∫–∏–ª–ª–æ–≤, –∑–∞—Ä–ø–ª–∞—Ç—ã
- –ö–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏—è –Ω–∞–∑–≤–∞–Ω–∏–π –≤–∞–∫–∞–Ω—Å–∏–π –∏ —Ä–∞—Å—á—ë—Ç —á–∞—Å—Ç–æ—Ç—ã

```python
import os
import time
import re
import pandas as pd
from dotenv import load_dotenv
from openai import OpenAI, OpenAIError
from tqdm import tqdm
from collections import Counter
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
from difflib import get_close_matches

# –ó–∞–≥—Ä—É–∑–∫–∞ API-–∫–ª—é—á–∞
load_dotenv()
api_key = os.getenv("OPENAI_API_KEY")
client = OpenAI(api_key=api_key)

# –§—É–Ω–∫—Ü–∏—è 1: –ö–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏—è –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏—è (Marketplace, Marketing, Sales)
def classify_direction(text: str) -> str:
    prompt = (
        "–¢—ã –∞–Ω–∞–ª–∏—Ç–∏–∫ –≤–∞–∫–∞–Ω—Å–∏–π. –ü–æ –æ–ø–∏—Å–∞–Ω–∏—é –≤–∞–∫–∞–Ω—Å–∏–∏ –æ–ø—Ä–µ–¥–µ–ª–∏, –∫ –∫–∞–∫–æ–π –≥—Ä—É–ø–ø–µ eCommerce –æ–Ω–∞ –æ—Ç–Ω–æ—Å–∏—Ç—Å—è: "
        "Marketplace, Marketing –∏–ª–∏ Sales. –û—Ç–≤–µ—Ç –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å —Ç–æ–ª—å–∫–æ –æ–¥–Ω–æ —Å–ª–æ–≤–æ ‚Äî –Ω–∞–∑–≤–∞–Ω–∏–µ –≥—Ä—É–ø–ø—ã.\n\n"
        f"–û–ø–∏—Å–∞–Ω–∏–µ:\n{text}"
    )
    try:
        response = client.chat.completions.create(
            model="gpt-4o",
            messages=[{"role": "user", "content": prompt}],
            temperature=0.2
        )
        return response.choices[0].message.content.strip()
    except Exception as e:
        print("‚ùå –û—à–∏–±–∫–∞:", e)
        return "Error"

# –§—É–Ω–∫—Ü–∏—è 2: –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ eCommerce –∏–ª–∏ offline
def classify_ecommerce_or_offline(text: str) -> str:
    prompt = (
        "–¢—ã –∞–Ω–∞–ª–∏—Ç–∏–∫ –≤–∞–∫–∞–Ω—Å–∏–π. –û–ø—Ä–µ–¥–µ–ª–∏, –æ—Ç–Ω–æ—Å–∏—Ç—Å—è –ª–∏ –≤–∞–∫–∞–Ω—Å–∏—è –∫ —Å—Ñ–µ—Ä–µ eCommerce –∏–ª–∏ –∫ –æ—Ñ—Ñ–ª–∞–π–Ω-—Ç–æ—Ä–≥–æ–≤–ª–µ. "
        "–û—Ç–≤–µ—Ç –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å —Ç–æ–ª—å–∫–æ –æ–¥–Ω–æ —Å–ª–æ–≤–æ: eCommerce –∏–ª–∏ offline.\n\n"
        f"–û–ø–∏—Å–∞–Ω–∏–µ:\n{text}"
    )
    try:
        response = client.chat.completions.create(
            model="gpt-4o",
            messages=[{"role": "user", "content": prompt}],
            temperature=0.2
        )
        return response.choices[0].message.content.strip()
    except Exception as e:
        print("‚ùå –û—à–∏–±–∫–∞:", e)
        return "Error"

# –§—É–Ω–∫—Ü–∏—è 3: –ò–∑–≤–ª–µ—á–µ–Ω–∏–µ —Ö–∞—Ä–¥/—Å–æ—Ñ—Ç —Å–∫–∏–ª–ª–æ–≤ –∏ –∑–∞—Ä–ø–ª–∞—Ç—ã
def extract_skills_and_salary(text: str) -> dict:
    prompt = (
        "–¢—ã –∞–Ω–∞–ª–∏—Ç–∏–∫ –≤–∞–∫–∞–Ω—Å–∏–π. –ü—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä—É–π –æ–ø–∏—Å–∞–Ω–∏–µ –∏ –∏–∑–≤–ª–µ–∫–∏:\n"
        "Hard skills: ...\nSoft skills: ...\nSalary: ...\n\n"
        f"–û–ø–∏—Å–∞–Ω–∏–µ:\n{text}"
    )
    try:
        response = client.chat.completions.create(
            model="gpt-4o",
            messages=[{"role": "user", "content": prompt}],
            temperature=0.3
        )
        output = response.choices[0].message.content.strip()
        lines = output.splitlines()
        result = {"hard_skills": "", "soft_skills": "", "salary_info": ""}
        for line in lines:
            if line.lower().startswith("hard skills:"):
                result["hard_skills"] = line.partition(":")[2].strip()
            elif line.lower().startswith("soft skills:"):
                result["soft_skills"] = line.partition(":")[2].strip()
            elif line.lower().startswith("salary:"):
                result["salary_info"] = line.partition(":")[2].strip()
        return result
    except Exception as e:
        print("‚ùå –û—à–∏–±–∫–∞:", e)
        return {"hard_skills": "Error", "soft_skills": "Error", "salary_info": "Error"}

# –û—Å–Ω–æ–≤–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –æ–±—Ä–∞–±–æ—Ç–∫–∏ –≤–∞–∫–∞–Ω—Å–∏–π
def process_vacancies(df: pd.DataFrame, pause_sec: float = 2.0) -> pd.DataFrame:
    direction_list = []
    ecommerce_list = []
    hard_skills = []
    soft_skills = []
    salary_info = []

    text_columns = df.select_dtypes(include=["object"]).columns.tolist()

    for _, row in tqdm(df.iterrows(), total=len(df), desc="–û–±—Ä–∞–±–æ—Ç–∫–∞ –≤–∞–∫–∞–Ω—Å–∏–π"):
        full_text = " ".join(str(row[col]) for col in text_columns if pd.notnull(row[col]))

        #direction = classify_direction(full_text)
        #ecommerce_type = classify_ecommerce_or_offline(full_text)
        skills = extract_skills_and_salary(full_text)

        #direction_list.append(direction)
        #ecommerce_list.append(ecommerce_type)
        hard_skills.append(skills["hard_skills"])
        soft_skills.append(skills["soft_skills"])
        salary_info.append(skills["salary_info"])

        time.sleep(pause_sec)

    #df["direction_llm"] = direction_list
    #df["ecommerce_or_offline"] = ecommerce_list
    df["hard_skills"] = hard_skills
    df["soft_skills"] = soft_skills
    df["salary_info"] = salary_info
    return df

# –ö–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏—è –Ω–∞–∑–≤–∞–Ω–∏–π –≤–∞–∫–∞–Ω—Å–∏–π
def cluster_job_titles(df: pd.DataFrame) -> pd.DataFrame:
    def clean_title(title):
        title = str(title).lower()
        title = re.sub(r"[^a-zA-Z0-9\s]", " ", title)
        title = re.sub(r"\s+", " ", title).strip()
        return title

    def remove_stopwords(text):
        return " ".join([word for word in text.split() if word not in ENGLISH_STOP_WORDS])

    df["job_title_clean"] = df["job_title"].apply(clean_title).apply(remove_stopwords)
    title_counts = Counter(df["job_title_clean"])
    total = sum(title_counts.values())

    def cluster_titles(titles, threshold=0.8):
        clusters = {}
        for title in titles:
            found = False
            for key in clusters:
                if get_close_matches(title, [key], n=1, cutoff=threshold):
                    clusters[key].append(title)
                    found = True
                    break
            if not found:
                clusters[title] = [title]
        return clusters

    clusters = cluster_titles(title_counts.keys())
    cluster_summary = []
    for cluster_name, variants in clusters.items():
        count = sum(title_counts[v] for v in variants)
        percent = round(count / total * 100, 2)
        cluster_summary.append({
            "cluster_name": cluster_name,
            "variants": ", ".join(variants),
            "count": count,
            "percent": percent
        })

    return pd.DataFrame(cluster_summary).sort_values(by="count", ascending=False)

# –ó–∞–ø—É—Å–∫ –∞–Ω–∞–ª–∏–∑–∞
if __name__ == "__main__":
    df = pd.read_excel("data/combined_ecommerce_jobs.xlsx")
    df = process_vacancies(df, pause_sec=1.0)
    df.to_excel("data/enriched_ecommerce_jobs.xlsx", index=False)
    print("–í–∞–∫–∞–Ω—Å–∏–∏ –æ–±—Ä–∞–±–æ—Ç–∞–Ω—ã –∏ —Å–æ—Ö—Ä–∞–Ω–µ–Ω—ã.")

    # cluster_df = cluster_job_titles(df)
    # cluster_df.to_excel("data/job_title_clusters.xlsx", index=False)
    # print("–ö–ª–∞—Å—Ç–µ—Ä—ã –Ω–∞–∑–≤–∞–Ω–∏–π —Å–æ—Ö—Ä–∞–Ω–µ–Ω—ã.")
```
