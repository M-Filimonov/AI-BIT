```python
import pandas as pd
from pathlib import Path

# Загружаем оба файла
file1 = Path("data") / "dataset_linkedin-jobs-scraper-no-login-required_2025-12-14_20-55-54-808.xlsx"
file2 = Path("data") / "dataset_linkedin-jobs-scraper-no-login-required_2025-12-14_21-42-36-690.xlsx"

df1 = pd.read_excel(file1)
df2 = pd.read_excel(file2)

print(f"Начальное количество записей df1: {len(df1)}")
print(f"Начальное количество записей df2: {len(df2)}")

# Объединяем по колонкам (выравнивание по именам)
df = pd.concat([df1, df2], ignore_index=True)

print(f"Итоговое количество записей после объединения: {len(df)}")

# Сохраняем в новый Excel
output_file = Path("data") / "dataset_linkedin-jobs-scraper-no-login-required_2025-12-14.xlsx"
df.to_excel(output_file, index=False)

print(f"Объединённый файл сохранён: {output_file}")
```
